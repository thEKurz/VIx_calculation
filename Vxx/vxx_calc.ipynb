```python
import pandas as pd
import numpy as np
import datetime as dt
```

```python
data = pd.read_excel("VIX_Futures_2022_1104.xlsx", index_col='Dates')
```

```python
data
```

```python
tbr_list = []
for p in range(0,len(data)):
    if p > 0:
        delta_t = data.index[p].day - data.index[p-1].day
        tbar_t_1 = data['1M Tbill'][p-1]
        tbr = (1/(1-91/360*tbar_t_1))**(delta_t/91)
    else:
        tbr=np.nan
    tbr_list.append(tbr)
```

```python
tbr_list
```

```python
cdr_list = []
e={}
e[0] = dt.datetime(2022,8,17)
e[1]= dt.datetime(2022,9,20)
e[2] = dt.datetime(2022,10,19)
e[3] = dt.datetime(2022,11,16)
e[4] = dt.datetime(2022,12,21)
f={}
f[1] = data['UXU2 Index']
f[2] = data['UXV2 Index']
f[3] = data['UXX2 Index']
f[4] = data['UXZ2 Index']
for p in range(0,len(data)):
    if p > 0:
        date = data.index[p]
        date_1 = data.index[p-1]
        if date < e[1]:
            n =1
        elif date < e[2]:
            n = 2
        else:
            n=3
        d_t = (e[n] - e[n-1]).days
        if p==len(data)-1:
            d_r = (e[n] - dt.datetime(2022,11,7)).days
        else:
            d_r = (e[n] - data.index[p+1]).days
        w_1 = d_r/d_t
        w_2 = 1 - w_1
        num = (w_1 * f[n].loc[date] + w_2 * f[n+1].loc[date])
        den = (w_1 * f[n].loc[date_1] + w_2 * f[n+1].loc[date_1])
        cdr = num/den
        cdr_list.append(cdr)
    else:
        cdr_list.append(np.nan)
```

```python
cdr_list
```

```python
data['tbr'] = tbr_list
data['cdr'] = cdr_list
data
```

```python
data['my_spv'] = (data['tbr'] + data['cdr']-1)* data['SPVXSTR Index'].shift()
```
